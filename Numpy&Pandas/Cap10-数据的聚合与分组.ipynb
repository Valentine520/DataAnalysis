```python
def draw(deck, n=5):
    if n > len(deck):
        print('Error n value')
        return None
    return deck.sample(n) # 抽样出5张图片
```

```python
import statsmodels.api as sm
```

```python
def regress(data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept'] = 1.
    result = sm.OLS(Y, X).fit()
    return result.params
```

```python
by_year.apply(regress, 'AAPL', ['SPX'])
```
